# Defining Hypotheses
## LESSON OBJECTIVES
By the end of this lesson, students will be able to:

- Define a Null Hypothesis and an Alternate Hypothesis
- Give examples of each

Before performing a statistical test, we need to define what we are testing.  There is a formal method for defining our test that involves writing two opposing statements that represent the two possible outcomes of our test. These are known as the null and alternate hypotheses.  We will explore these terms with an example.

## Null and Alternate Hypotheses
Scenario: A new medicine is being developed for high blood pressure, and its effects on a patient's health are being tested in a clinical trial. There are lots of considerations for designing a clinical test (which can be very interesting!) , but we are focused on understanding how we might define the null and alternate hypothesis.  Assume a test group of patients are given the new medicine, while others (in a control group) are given a placebo.



### Null Hypothesis (no difference)

The null hypothesis () is a statement that essentially indicates nothing notable has been found.  More specifically, it proposes that NO statistical significance exists.  It is sometimes considered the statement of what would be common or already accepted.  It can also be the statement that no difference exists.

- The null hypothesis would essentially be that the medicine has no effect on blood pressure.
    - Null hypothesis: There is no difference between the outcomes of patients who took the new medicine. vs. those who took the placebo.


### Alternate Hypothesis (difference)

The alternate hypothesis ( ) is what the researcher/data science is trying to test. It indicates that something notable has been been found. It proposes that there IS a statistically significant finding.  Note that usually, researchers/data scientists are hoping to support the alternate hypothesis as it would mean they have a finding that is different to what is commonly accepted or they have shown a difference exists that may be notable.

- The alternate hypothesis would be that the medicine did have an effect on blood pressure.
    - Alternate hypothesis: There is a difference between the outcomes of patients who took the new medicine vs. those who took the placebo.

Note that in some cases, we are just looking for any difference, but in other cases we are looking for a difference in a particular direction.  For example, in the case of our new medicine, we may want to further specify what we mean by "outcome " in the alternate hypothesis.  In this case, we are trying to test if our new medicine lowers high blood pressure. 

- Refined alternate hypothesis: The blood pressure of patients taking the new medicine is lower than those who took the placebo.


## Scenario: Are they part of the human population?


- Let's assume, for a moment, that an alien species has begun infiltrating humanity, disguised as a typical human.
    - They look VERY much like a human, but they are known to be taller than the vast majority of humans.
- We have discovered an individual that we suspect to be alien, but we need some way to mathematically support our belief that this individual is not part of the human population (before we act on it).  You very carefully/secretly measured their height, which was 78 inches tall.
- You have access to data for human height. 

## Defining Our Hypotheses
- We now have a belief that we want to test and support with statistical analysis.
- We want to find out if our unknown lifeform's height is significantly taller than the human population.
- What do you think would be appropriate null  and alternate hypotheses for this test??

### Stating Our Hypothesis and Null Hypothesis

Belief/suspicion that we want to test:  Is this potential alien's height a normal height for a human or is it significantly taller?

- : Null Hypothesis "The potential alien's height is from the human population."
- : Alternate Hypothesis: "The potential alien's height is significantly larger than the human population"

## Outcomes
After performing the test, we will have one of two outcomes. These are typically defined in terms of whether or not we rejected the null hypothesis.  

You will often see the following formal language used in association with statistical testing

As a result of testing, you either:

- "Reject" the null hypothesis 
                         OR

- "Fail to reject" the null hypothesis.


Sometimes this formal language can be confusing! It will get easier with experience, but let's translate these:

- "Reject" the null hypothesis means you have found a "real" (statistically significant) difference. 
    - You have supported the alternate hypothesis. 


- "Fail to reject" the null hypothesis means you must continue to accept the null hypothesis.  You have not provided any support for the alternate hypothesis.
    - You must reject the alternate hypothesis.


## Summary
A first step in performing a statistical test is to write two clear statements that represent the two possible outcomes of your test.  These statements are always paired and are always opposite of each other. The null hypothesis states that there is no difference, while the alternate hypothesis states there is a difference.  After performing the test, you will either either:

- reject the null hypothesis (this means you supported your alternate hypothesis)
- accept ("fail to reject")  the null hypothesis (this means you did NOT support your alternate hypothesis)

In the next lesson, we will explore how we determine whether we accept or reject our null hypothesis.
