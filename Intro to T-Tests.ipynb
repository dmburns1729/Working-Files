## When to use a t-test?
One of the most common uses of a t-test is to determine whether the means of two groups are different.

There are tons of examples where two groups might be compared: Just a few examples include:

- Are the heights of males and females different?
- Are the star-ratings of two brands of shoes different?
- Is the number of laughs occurring between a movie and its sequel different?
- Is the resting heart rate lower after an exercise program?
- Is the number of subscriptions obtained from a website with a red subscribe button higher than a website with a green subscribe button?

Note that there almost always IS some difference in the means of the groups, but the t-test is important because it determines if that difference is meaningful or if it is just due to randomness!

The exact question you want to answer will help you decide if a t-test is the appropriate option, and will help you determine which of several t-tests you should use.  The main types of t-tests are 

- independent (between groups) -when different participants are in each group (such as heights of males and females) 
    - scipy.stats.ttest_ind (documentation)
- dependent (repeated measures, "paired t-test" - often considered "pre/post" and is used when the same participants are in both groups (such as the resting heart rates of the same participants measured before and after the exercise program)
    - scipy.stats.ttest_rel (documentation)
- one sample t-tests compare the mean of one group to a well established mean
    - scipy.stats.ttest_1samp (documentation)

The null hypothesis for any t-test is that the means are the same. In other words, there is no difference.

The result of a t-test, like most statistical tests, will be a single value. In this case it is called the "t-statistic" or "t-score".  The t-score is a ratio between the difference between two groups and the difference within the groups. There are ways to calculate this manually, but we will let scipy do this for us!

You will also get a value for p which you will interpret as you do any p value (and as we did for our normality tests).  You determine the alpha, such as 0.05 and if you get a p value less than 0.05, you can reject the null hypothesis and conclude that you have a meaningful difference between your groups.  

We will work through an example scenario to demonstrate the application of a t-test.





## Example: Question: do males and females have different average heights?

Previously, we have been referring the Male vs Female Height & Weight data as our population. Truthfully, it is actually just a VERY large sample (N=10,000 males & females). 




- Notice, that while the two means are pretty far apart, there is a lot of overlap between the two distributions.

## A More Realistic Sample Size (N=50)

Note: when we have 2 groups, we will use 2 different representations of n.
- Capital N: the total number of observations across all groups.
- Lowercase n: the number of observations in 1 group. (note that the groups do not need to be of equal size)

```python
#Imports

import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
#Load and Filter Data

#The code below creates a random sample of 50, and splits the data based on male or female. The length and mean of each group is printed.

df = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vRxp-hsw61ck9cpt5ExZIUhvSGT5zYzoUJl_7xr7YF5qZeazC8393q2taqwkDvMxVirgFDJm1WX3uAC/pub?output=csv")
sample = df.sample(n=50,random_state=42)
female_height = sample.loc[ sample['Gender']=='Female',"Height"]
male_height = sample.loc[ sample['Gender']=='Male',"Height"]
## Getting means for gender height
print(f"For Males (n={len(male_height)}): Mean={np.mean(male_height):.2f}")
print(f"For Females (n={len(female_height)}): Mean={np.mean(female_height):.2f}")
```

You can see above that there are two groups. There are 27 males with an average height of 68.90 and 23 females with an average height of 64.13.

### Hypothesis 

As you can see, the means are clearly different, but we will use the t-test to determine if this difference is statistically significant or if it is just due to randomness in our samples.

Before we can begin testing our assumptions, we must formally declare our null and alternative hypothesis and set our alpha level. 

Null Hypothesis: Males and females have the same average height. (There is no statistical significance between the heights of the groups)

Alternative Hypothesis: Males and females have different average height. (There is a statistical significance between the heights of the two groups)

Alpha: We will be using an alpha=.05

## Test Assumptions 
First, we recognize that males and females must be from different groups, so we will want to perform an independent t-test.

Just because we WANT to perform an independent t-test (or any test), doesn't mean that the test is appropriate for our data.   Each statistical test has some assumptions that must be met in order to trust the resulting test statistic and p-value.  By building our background knowledge of distributions throughout the week, we have been building up to this point where you can better understand such assumptions!

There are three main assumptions that we must test before trusting the results of an independent t-test. 

Independent T-Test Assumptions:

- No significant outliers
- Normality 
- Equal Variance 

### Testing for Outliers

There are several ways to test for outliers, but z-scores are a very simple way to identify if you have outliers and which values they are. Recall, that a z-score is how many standard deviations a value is from the mean.  Typically z-scores greater than 3 are considered outliers. 

```python
zscores_f = stats.zscore(female_height)
outliers_f = abs(zscores_f)>3
np.sum(outliers_f)
```

```python
zscores_m = stats.zscore(male_height)
outliers_m = abs(zscores_m)>3
np.sum(outliers_m)
```

There are no outliers in our data.

If outliers are present, you would want to remove them.

```python
# Test Each Group for Normality
# We have used the test for normality in previous lessons. 
# We will apply it here to test each of our groups separately. 

# normal test males
result_m = stats.normaltest(male_height)
result_m
```

```python
# normal test females
result_f = stats.normaltest(female_height)
result_f
```

For both the male and female group, the p-value is greater than 0.05 so we can assume these are normal distributions.

If the normality test is NOT met, you can proceed if there is a large sample.

For a two sample t test, if each group is 15, the sample size is considered large enough to proceed.

## Testing for Equal Variance with Levene's test
Equal variances means that the variance is approximately the same across the samples.  This is also known as the assumption of homoscedasticity or the homogeneity of variances.   Just like with normality, it can be assessed with a simple test as long as you know which test and you know how to interpret p-values!  The test we will use is called Levene's test.  Click here to explore the
documentation

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.levene.html#scipy.stats.levene

As with any statistical test, you must know what the null hypothesis is so you are able to interpret the results.  For Levene's test, the null hypothesis is: All samples are from populations with equal variances.  

```python
result = stats.levene(female_height, male_height)
result
```

```python
#Independent t-test
#Click here to explore the documentation for this test.



## Final t-test, after confirming we meet the assumptions
result = stats.ttest_ind(female_height,male_height)
result
```

```python
## is our result significant
print(f"p-value={result.pvalue:.10f}")
print(f"Significant: {result.pvalue <.05}")
```

Based on our t-test, we have a significant result. Remember that a significant result (p < 0.05) means we reject the null hypothesis and accept the alternative hypothesis.  In this case, this means that there IS a difference between male and female heights.

In conclusion, we performed an independent t-test comparing height by gender and found a significant difference between males and females (p<.05). Males had a significantly larger height than females.
