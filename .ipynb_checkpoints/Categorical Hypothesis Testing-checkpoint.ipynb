## Tests for Categorical Targets

- So far, we have only covered tests that compare a target numeric feature across groups/samples.
    - Sometimes, the groups themselves are the target feature.
- In this lesson we will cover 2 types of categorical hypothesis tests:
    - Binomial test:
        - Testing the proportion of binary outcomes (Wins/Losses, True/False) vs. the expected probability/proportion.
        - Example: Testing if a coin is a fair coin by flipping it 10 times.
    - Chi-squared () test:
        - Testing Group Membership Across Categories.
        - Example: Testing if there is a difference in the amount of Smokers/Non-Smokers between Gender (Male/Female).

## Binomial Test
- For comparing binary outcomes vs. an expected probability of success.
- Null Hypotheses:
        - That the outcomes occur at the expected probability.
        - Example: the coin is a fair coin (probability of heads=.5)
- Binomial Test Assumptions
        - There are 2 possible outcomes: success and failure. 
        - The probability of success is constant.
        - The trials are independent (outcome of one trial does not influence another).

## Using scipy.stats.binom_test
Documentation

The null hypothesis of the binomial test is that the probability of success is the expected probability.

Success and failure do not necessarily indicate a desirable or non-desirable outcome. In many cases, it is arbitrary which outcome you define as "success".  For example, in a coin toss, we can define success as "heads", but it certainly doesn't mean that heads is a better outcome than tails.  As with many cases in statistical analysis, it is just essential that you are clear in how your scenario is defined.

- Parameters:

        - x: int or array_like

            - The number of successes, or if x has length 2, it is the number of successes and the number of failures.
        - n: int
            - The number of trials. This is ignored if x gives both the number of successes and failures.
    - p: float, optional

            - The hypothesized probability of success. 0 <= p <= 1. The default value is p = 0.5. (We would expect the outcome to be 50/50, or there is a 50% chance of each outcome)  Note that this "p" is NOT the p-value associated with significance!
    - The result of this test is the p-value.

## Example: Is this a fair coin?
### Coin A
You want to test if Coin A is a fair/balanced coin.

    - You flipped the coin 20 times.

        - The outcomes were:
            - 14 Heads
            - 6 Tails
    - Is coin A a fair coin? (Did it have a .5 probability of heads?)

    - Hypothesis:

        - Null Hypothesis: the probability of heads = .5
        - Alternative Hypothesis: the probability of heads is different than .5
        
        
        
 

```python
#Imports

import scipy.stats as stats
```

```python
# Save outcomes of 14 H (success), 6 T (failure)
outcomes = [14,6]
p_value = stats.binom_test(outcomes,p=0.5) # p is the expected probability of success (heads)
p_value
```

Since p>.05, we failed to reject the null hypothesis that the coin has a .5 probability of heads. We accept the null hypothesis that the probability of heads is .5 and can conclude that Coin A is a fair coin.

## Coin B
Now, you want to test if Coin B is a fair/balanced coin.

    - You flipped the coin 20 times.

Hypothesis:

## Chi-Squared
- For comparing a binary target category vs categorical groups.

- Null Hypothesis:

    - There is no relationship between the distribution of the target category and the categorical features.
- Chi-Square test

    - There are two categorical variables (ordinal or nominal)
    - The outcomes are independent

## Using scipy.stats.chi2_contingency
Documentation

"Chi-square test of independence of variables in a contingency table."

"This function computes the chi-square statistic and p-value for the hypothesis test of independence of the observed frequencies in the contingency table observed."

The input for the chi square test is a contingency table, which can also be known as a "crosstab." Below is an example of a contingency table for survival by sex on the Titanic created using pd.crosstab().





## Example: Titanic Survivors by Sex
- You want to test if there is a difference between the number of Men and Women who Survived/Died on the Titanic?

- Hypothesis:

    - Null Hypothesis: There is no relationship between Sex and if someone Survived or Died.
    - Alternative Hypothesis: A passenger's Sex influenced if they Survived or Died.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vRsXeEswMPJqEh9xpXed0eJYaQf_aKkNCypU4TKvGrS_hucLW2IWUxrVBjlKQJR4Z_EQFE-YR4UUuTz/pub?output=csv")
df.head(3)
```

```python
# First, let's visualize survivorship by sex

ax = sns.countplot(data=df, x='Sex',hue='Survived',)
ax.set_title("Number of Survivors by Sex");
```

## Getting the Contingency Table with pd.crosstab
- In order to run the Chi-Squared Test, we need a contingency table.
    - In other words, we need to know how many observations from each group variable (Sex) fell into which category in the target (Survived).
- We can use pd.crosstab to calculate this table.

```python
# calculate the contingency table with pd.cross-tab
table = pd.crosstab(df['Sex'], df['Survived'])
table
```

```python
# Run the Test and Save the Output
result = stats.chi2_contingency(table)
chi2, p, deg_free, expected_vals = result #this labels the results
```

```python
# What was our p-value?
print(p)
p<.05
```

The p-value is much less than 0.05 (always be careful to note the scientific notation "e-58"), so we have a significant result.



Since our p < .05, we reject the null hypothesis that there is no relationship between a passenger's Sex and if they Survived.
This means that we accept the alternate hypothesis and conclude that there is a relationship between a passenger's sex and survival.



## Summary
This lesson introduced two additional statistical tests that you can add to your growing toolbox of tests.  The binomial test determines if the outcome is statistically different than what would be expected based on probability when there are two possible outcomes. The input for a binomial test is the number of "successes" and "failures" as arbitrarily defined by the researcher.  A chi-squared test is used to determine if there is a relationship between categorical features. The input of the test is a contingency table which shows the frequency of each outcome  for each group. The contingency table can easily be created using pd.crosstab(). 

Just as with other statistical tests, you must know how the the null hypotheses are defined so you can appropriately interpret the meaning of a significant result.
