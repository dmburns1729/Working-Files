```python
import json
with open('/Users/dmburns/.secret/yelp_api.json') as f:
    login = json.load(f)
login.keys()
```

```python
# import the YelpAPI Class
from yelpapi import YelpAPI
# Create an instance with your key
yelp_api = YelpAPI(login['api-key'], timeout_s=5.0)
yelp_api
```

```python
# use our yelp_api variable's search_query method to perform our API call
search_results = yelp_api.search_query(location='Culver City, CA',
                                       term='Irish Pub')
print(type(search_results))
search_results.keys()
```

```python
search_results['total']
```

```python
import pandas as pd

irish_pubs_cc = pd.DataFrame(search_results['businesses'])
irish_pubs_cc.head(2) 
```

```python
## how many businesses in our results
len(search_results['businesses'])
```

```python
search_results = yelp_api.search_query(location='Culver City, CA',
                                       term='Irish Pub',
                                       offset = 20)
```

```python
irish_pubs_cc_20 = pd.DataFrame(search_results['businesses'])
irish_pubs_cc_20.head(2)
```

```python
## concatenate the previous results and new results. 
irish_pubs_cc = pd.concat([irish_pubs_cc, irish_pubs_cc_20],
                      ignore_index=True)
display(irish_pubs_cc.head(3), irish_pubs_cc.tail(3))
```

```python
irish_pubs_cc.to_json(orient='records')
```
