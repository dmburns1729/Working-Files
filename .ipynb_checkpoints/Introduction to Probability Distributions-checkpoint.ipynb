The y-axis on a histogram provides the actual raw count (frequency) of values that fall into each bin.  We are now going to convert these histograms to show the probability that a randomly selected value would fall into each particular bin.

First, this is the code to make a typical histogram (also known as a frequency histogram) with the count on the y-axis.  

Notice that we added discrete=True when creating a histogram for “Outlet_Establishment_Year”. This is because the Year is a discrete distribution and we want to visualize each year as a separate bar.
```python
sns.histplot(data=df, x='Outlet_Establishment_Year', discrete = True);
```

Probability Mass Function
The histogram above plots the actual count of values that fall into each bin/bar.
By dividing the raw counts by the overall number of values (n), we can convert the raw counts per bin into the probability that a value will fall into that bin.  

A probability can easily be converted into a percentage by multiplying the probability by 100.  For example, a probability of 0.10 is the same as 10%.

You can achieve this by adding stat = "density" when you make the histogram.  This is a visual representation of the Probability Mass Function (PMF) for Outlet_Establishment_Year.

```python
sns.histplot(data=df, x='Outlet_Establishment_Year', discrete=True, stat='density');
```

The only thing that has changed is the units of the y-axis.
It converted the raw # of counts to a fraction out of the total number of values.
When we convert a histogram of raw counts into the % of values that fell into each bin, we have created what is called a "Probability Mass Function". (PMF).
Note: the word "function" is used a little differently in statistics, as it means any equation that takes an input and returns an output.

Getting Probabilities from a PMF
A PMF is a discrete (blocky, non-continuous) probability distribution.
The sum of all of the bars = 1.  (A probability of 1 is equivalent to 100%)
We can determine the probability of a certain value occurring by looking at the height of the corresponding bin.

### From Discrete PMF to Continuous Kernal Density Estimates (KDE)
In order for us to get estimated probabilities for new values that were not in the original PMF, we can convert our PMF's bars into a continuous line by converting our discrete probability mass to a continuous probability density.

We can estimate the continuous probabilities by transforming our discrete PMF bars into a smoothed curve known as a Kernal Density Estimate.

This estimated probability curve is called a kernel-density estimate (KDE). Basically, a KDE is a way to interpolate between histogram bars so that we can get probability estimates for any value in and around our original distribution. 

```python
# The code to produce the KDE is sns.kdeplot:

sns.kdeplot(data = df, x= "Outlet_Establishment_Year");
```

```python
sns.histplot(data=df, x='Outlet_Establishment_Year', discrete=True, stat='density', kde = True);
```

## Summary
In summary, the probability mass function is simply the counts of our histogram divided by the total number of data points.

- Now, the values no longer tell us how many values fall in a bin. It now tells us the probability of a randomly selected row/data point falling into that bin.
- We can estimate the probabilities of new data using the continuous kernel density estimate (KDE).

### How the PMF and KDE are Used in Data Science
- Once we are able to get a continuous probability curve for distributions, we are able to calculate probabilities of seeing a specific value when we include a new data point.

## What are Measures of Central Tendency?
A measure of central tendency is a single value that attempts to describe a set of data by identifying the central position within that set of data.  In statistics, several different measurements are used to describe the “center” of data. These are mean, median, and mode. Using these three main measures, it is possible to find the central or typical value for a set of data. We can also also apply these measures to find the mean, median, and mode of a probability distribution. 



### Mean
The Mean or Arithmetic Average is the value obtained by dividing the sum of all the data by the total number of data points as shown in the formula below:

Means are used most commonly in the general public. However, the mean is easily influenced by outliers and is therefore not necessarily the best measure of central tendency.





### Median

The median is the middle-most value in a sorted list of all values/observations.  Note: If there are an even number of values, take the mean of the 2 middle values to get the median.



### Mode

The mode is the value that occurs the most frequently.
Note: it is possible for there to be more than 1 mode. In this case, it would be a multi-model distribution.

## How Are They Used?
The mean is the most frequently used measure of central tendency and is generally considered the best measure of it. However, there are some situations where either median or mode are preferred.

The median is the preferred measure of central tendency when there are a few extreme scores (outliers) in the distribution of the data, or there is a long tailed distribution. A long tailed distribution occurs when there are extreme values on one side (in the graph above there is a long tail on the right side with values greater than the median).

The mode is the preferred measure when data are measured in a nominal (and even sometimes ordinal) scale. 

## Skew and Measures of Central Tendency
In a normal distribution, the mean, median, and mode all correspond to the same value.  This is considered symmetrical, and therefore there is no skew.

What Is Skew?
Skew is the measure of the asymmetry of a distribution in a data set.

- In a negatively skewed distribution, more of the values are grouped to the right.
    - A real-life example of negatively skewed data is the distribution of the age of deaths in most populations.
        - Most people live to be between 70 and 80 years old, with fewer and fewer living less than this age.
- In a positively skewed distribution, more of the values are grouped to the left.
    - A real-life example of positively skewed data is income distribution.
        - A large percentage of the total people residing in a particular area tend to be low-income earners, while only a few people fall under the high-income earning group.

## Measures of Dispersion
A measure of dispersion is a single value that attempts to describe how far apart the values are compared to the center of the data.

The following is an example: the average data scientist salaries for 500 companies based on this kaggle dataset have been plotted against arbitrary values on the x-axis to show all of the values at once.

- The mean of the data is indicated by the dashed line.

### Understanding Variance

All measures of dispersion are interested in describing the typical distance between the mean and any data point 

               
We want to summarize the total amount of separation for all data points: 

                       
However, if we simply took the sum of all the differences, we would have some positive values for points above the mean, and some negative values for points that were below the mean.

Since we want to capture the typical spread of all data points, we do not want positive and negative values to cancel each other out!

To avoid this, we can square each value before summing them, since squaring a negative value will result in a positive one.
                    
We then, finally, divide by the number of values n to get our final equation.

This is what we call variance, which we represent mathematically as

The major disadvantage to variance, however, is that the measure of dispersion is now in different units than the central measure and other data points!

In our example, our units are thousands of dollars,  but the variance would be in thousands of dollars-squared! Since working with squared units is not practical, we simply take the square root of the variance to get the commonly used measure known as standard deviation. 



### Standard Deviation
Standard deviation is based on variance. The benefit of using standard deviation is that the units will be in the same units of the data set.

Calculating standard deviation 

The Standard Deviation is another measure of the spread of values within a dataset. It is simply the square root of the variance. 

Notice, that the symbol for standard deviation is just sigma: 

Thankfully, we can do use python for our calculations!  Standard deviation is used similarly to other calculations we have done such as .mean(), .max(), etc.

```python
# to calculate the standard deviation of a column in our dataset
df['Column_name'].std()
```

### Understanding standard deviation

Greater standard deviation indicates that the distribution of data points is more spread out away from mean.  The distribution graph will be wider.  Smaller standard deviation indicates that the distribution of data points is less spread out away from mean. The distribution graph will be thinner. Note that standard deviation can never be negative, but (in rare cases) it could be zero. If all of the values were identical, there would be a standard deviation of zero. (All of the values would be equal to the mean). The image below shows an example of two distributions with the same means but different standard variations.

The graph below shows the salary values as before, but this time, the green areas indicate indicating values within +1 and -1 standard deviations from the mean.

A more common way to visualize the distribution is with a histogram. You can see that the graph below shows values within 1 standard deviation (green) and within 2 standard deviations (yellow). Values beyond the colored area are more than 2 standard deviations away from the mean.  

Values within the green range are considered "within 1 standard deviation of the mean".   Data in this range are common and are not considered unusually high or low.  Data beyond 1 standard deviation are considered less common and is either high or low.  If a value is beyond 3 standard deviations from the mean, it is generally considered an outlier.

We have already starting talking in terms of "how many standard deviations." Notice that the number of standard deviations is informative even if we don't know the scale or units of the original data. You will continue to explore this concept by using z-scores.

### What Are Z-Scores?
Z-scores are another way of expressing how how much a value differs from the mean.  A z-score is measured in units of standard deviation.  A value that is 1 standard deviation away from the mean has a z-score of 1.  A value that is 2 standard deviations away from the mean has a z-score of 2. (And so on...)

Notice in the graph above that the only change is the units on the x-axis. Rather than being in the original units, they are in number of standard deviations (Z-scores).

### How is it Calculated?

Z score can be calculated by finding the difference between the data point and the mean, and dividing this value by the standard deviation.



For example lets say that a distribution has mean value of 500 and a standard deviation of 25.

What is the z-score for the value 525? 

(525-500)/25 = 1   (525 is within 1 standard deviation of the mean so it has a z-score of 1)

What is the z-score for the value 550?

(550-500)/25 = 2 (550 is within 2 standard deviations of the mean so it has a z-score of 2)



### Kurtosis
What Is Kurtosis? First, it is important to note that kurtosis is NOT a measure of variance. It does, however, describe a different aspect of the distribution.

Kurtosis is a statistical measure used to describe the degree to which the distribution of values cluster in the tails or the peak of a frequency distribution. The peak is the tallest part of the distribution, and the tails are the ends of the distribution. There are three types of kurtosis:

- Mesokurtic (values cluster in the middle of the distribution - normal)
- Leptokurtic ("pointy", values cluster in the peak of the distribution - positive)
- Platykurtic ("flat", values cluster in the tails of the distribution - negative).

A detailed explanation of kurtosis can be found here.

Thankfully, we do not have to eyeball kurtosis, as scipy's stats module has a kurtosis function.

To calculate the kurtosis, you can use the following code. We set fisher = False to use the Pearson calculation. The default is the Fisher calculation which uses a different scale for interpretation. See documentation here.

```python
from scipy import stats
kurt = stats.kurtosis(df['column_name'], fisher = False)
```

Using the Pearson calculation, 

Mesokurtic (Kurtosis ~3)

Leptokurtic (Kurtosis >3)     

Platykurtic (Kurtosis < 3)
